```python
import pandas as pd
sheet_id = "1gSflglwFiLXCpzflf8_2C6-DalZGI1AVRXfytZZtAYA"
sheet_name = "p_anax"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
genx = pd.read_csv(url, low_memory=False)
# df_main = df_main[df_main.tahun==2021]
genx = genx.iloc[:,1:77]
```

```python
# Encode categorical variables using one-hot encoding
encoded_data = pd.get_dummies(genx)

# Calculate the correlation matrix for all columns
corr_matrix = encoded_data.corr()

# Select the correlation values for the last 7 columns
last_7_corr = corr_matrix.iloc[-7:, :]

# Display the correlation values for the last 7 columns
# print(last_7_corr)

# # List the column names with the highest correlation for each column
highest_corr_columns = last_7_corr.idxmax()

# # Print the column names with the highest correlation
highest_corr_columns

# Create a DataFrame to display the pairs with the highest correlation
df_highest_corr_pairs = pd.DataFrame(highest_corr_pairs, columns=['Correlation'])

# Print the DataFrame
df_highest_corr_pairs
```

```python
# Assuming you have a DataFrame called genx
# genx = pd.DataFrame(...)

# Define the mapping dictionary
mapping = {
    'Strongly Agree': 5,
    'Agree': 4,
    'Neutral': 3,
    'Disagree': 2,
    'Strong Disagree': 1
}

# Iterate over each column from index 1 to 69 and apply the mapping
for column_index in range(0, 70):
    genx.iloc[:, column_index] = genx.iloc[:, column_index].map(mapping)
```

```python
# Calculate the maximum value by row
forsdata['max_value'] = forsdata.max(axis=1)

# Calculate the minimum value by row
forsdata['min_value'] = forsdata.min(axis=1)

forsdata['consistency'] = forsdata['max_value']/forsdata['min_value']
```

```python
import random
import pandas as pd
from fancyimpute import IterativeImputer

# Assuming you have already loaded your dataframe 'genx'

# Select relevant columns
# forsdata = genx.iloc[:, [i for i in range(0, len(genx.columns)) if i in [70, 71, 72]]]

# Randomly select values for 'Employment Sector', 'Education Level', and 'Monthly Income'
sector = random.choice(forsdata['Employment Sector'].unique())
education = random.choice(forsdata['Education Level'].unique())
income = random.choice(forsdata['Monthly Income'].unique())

# Fill the selected values in the dataframe
forsdata.loc[forsdata['Employment Sector'].isnull(), 'Employment Sector'] = sector
forsdata.loc[forsdata['Education Level'].isnull(), 'Education Level'] = education
forsdata.loc[forsdata['Monthly Income'].isnull(), 'Monthly Income'] = income

# Now, let's use MICE to impute missing values in other columns
exclude_columns = ['Employment Sector', 'Education Level', 'Monthly Income']

# Specify columns to impute
columns_to_impute = [col for col in forsdata.columns if col not in exclude_columns]

# Create imputer 
imputer = IterativeImputer()

# Fit and transform to impute 
forsdata_imputed = pd.DataFrame(imputer.fit_transform(forsdata[columns_to_impute]), 
                               columns=columns_to_impute)  

# Replace imputed values
forsdata[columns_to_impute] = forsdata_imputed[columns_to_impute]

# Generate 200 synthetic data points
synthetic_data = pd.DataFrame(columns=forsdata.columns)
for _ in range(200):
    synthetic_row = forsdata.iloc[random.randint(0, len(forsdata) - 1)]
    synthetic_data = synthetic_data.append(synthetic_row, ignore_index=True)

# Convert numeric columns to integers
synthetic_data[columns_to_impute] = synthetic_data[columns_to_impute].astype(int)

# Display the synthetic data
synthetic_data
```

```python
synthetic_data.to_csv('synthetic_data_genX.csv', index=False)
```

```python
synthetic_data.to_clipboard(index=False)
```
