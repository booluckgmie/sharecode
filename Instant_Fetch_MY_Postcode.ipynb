## one by one

```python
import requests
import json

poskod = '02600'
# Define the URL with the specific postcode
url = f"https://api.pos.com.my/PostcodeWebApi/api/Postcode?Postcode={poskod}"

# Make the request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = json.loads(response.content)

    # Check if there is at least one entry in the data
    if data:
        # Get the first entry
        first_entry = data[0]

        # Extract Post_Office and State from the first entry
        post_office = first_entry["Post_Office"]
        state = first_entry["State"]

        # Print Post_Office and State
        print(f'{post_office}, {state}')
    else:
        print("No data available")
else:
    print("Failed to retrieve data")
```

## In Bulk

```python
import pandas as pd

# Sample dataframe
data = {
    'no': [1, 2, 3],
    'poskod': [16600, 62050, '02600'],
    'city': ['??', '??','??']
}

df = pd.DataFrame(data)
df
```

```python
# Function to fetch data from API and match
def fetch_and_match_data(poskod):
    url = f"https://api.pos.com.my/PostcodeWebApi/api/Postcode?Postcode={poskod}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        if data:
            # Assuming we are only interested in the first entry
            first_entry = data[0]
            return f"{first_entry['Post_Office']}, {first_entry['State']}"
        else:
            return "No data available"
    else:
        return "Failed to retrieve data"

# Apply the function to each row in the dataframe
df['city'] = df['poskod'].apply(fetch_and_match_data)

# Print the resulting dataframe
print(df)
```
