```python
# Calculate the total number of genders
df_all['totalgender'] = df_all['male'] + df_all['female']

# Calculate the percentage of male and female, rounded to 1 decimal place
df_all['male'] = round((df_all['male'] / df_all['totalgender']) * 100, 1)
df_all['female'] = round((df_all['female'] / df_all['totalgender']) * 100, 1)
```

```python
# Calculate the total number of labour
df_all['totallabour'] = df_all['working'] + df_all['unemployed']  + df_all['out']

# Calculate the percentage of male and female, rounded to 1 decimal place
df_all['working'] = round((df_all['working'] / df_all['totallabour']) * 100, 1)
df_all['unemployed'] = round((df_all['unemployed'] / df_all['totallabour']) * 100, 1)
df_all['out'] = round((df_all['out'] / df_all['totallabour']) * 100, 1)
```

```python
df_all = df_all.drop(['totalgender', 'totallabour'], axis=1)
df_all
```

```python
import pandas as pd
import numpy as np

def convert_to_float(df, exclude=[]):
    # Select columns that contain numeric data
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    # Remove excluded columns from numeric columns list
    numeric_cols = [col for col in numeric_cols if col not in exclude]

    # Round all numeric columns to 1 decimal place
    df[numeric_cols] = df[numeric_cols].astype(float).round(1)

    return df
```
