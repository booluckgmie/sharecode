```python
import pandas as pd

URL_DATA = 'https://storage.googleapis.com/dosm-public-dashboards/kawasanku_admin_pyramid.parquet'
df = pd.read_parquet(URL_DATA)

# Convert date column to datetime if it exists
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])

# Define geo_levels to filter by
geo_levels = ['country', 'state', 'district']

for geo_level in geo_levels:
    # Filter data by geo_level
    df_geo = df[df['area_type'] == geo_level].reset_index(drop=True)
    
    # Generate CSV file for this geo_level
    filename = f'kawasanku_admin_pyramid_{geo_level}.csv'
    df_geo.to_csv(filename, index=False)
```
