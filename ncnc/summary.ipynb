```python
# Loop through each unique value in 'UPDATED_BY'
for user in df2['UPDATED_BY'].unique():
    # Filter data for the current user
    user_df = df2[df2['UPDATED_BY'] == user]
    
    # Clean filename (optional: remove or replace special characters)
    safe_user = str(user).replace(" ", "_").replace("/", "_")
    
    # Save to a new CSV file
    user_df.to_csv(f'dataMCMC_{safe_user}.csv', index=False)
```
