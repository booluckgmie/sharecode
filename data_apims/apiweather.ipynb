```python
import requests
import pandas as pd

# Define the URL
url = "https://eqmp.doe.gov.my/api2/publicportalapims/apitablehourly?datetime=2017-07-31"

# Fetch the data from the API
response = requests.get(url, verify=False)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()
    
    # Extract the relevant part of the JSON
    api_data = data['api_table_hourly']
    
    # Load the data into a DataFrame
    df = pd.DataFrame(api_data)
    
    # Display the DataFrame
    print(df)
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Define the start and end dates
start_date = datetime(2017, 7, 31)
end_date = datetime(2024, 11, 4)

# Create a list to hold the DataFrames for each date
dataframes = []

# Loop through each date in the range
current_date = start_date
while current_date <= end_date:
    # Format the date to match the URL requirement
    formatted_date = current_date.strftime('%Y-%m-%d')
    
    # Define the URL for the current date
    url = f"https://eqmp.doe.gov.my/api2/publicportalapims/apitablehourly?datetime={formatted_date}"
    
    # Fetch the data from the API
    response = requests.get(url, verify=False)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON data
        data = response.json()
        
        # Extract the relevant part of the JSON
        api_data = data['api_table_hourly']
        
        # Load the data into a DataFrame
        df = pd.DataFrame(api_data)
        
        # Append the DataFrame to the list
        dataframes.append(df)
    else:
        print(f"Failed to fetch data for {formatted_date}: {response.status_code}")
    
    # Move to the next date
    current_date += timedelta(days=1)

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dataframes, ignore_index=True)

# Display the final DataFrame
print(final_df)
```
