```python
import requests
from scidownl import scihub_download
from tqdm import tqdm
import os

# Make sure the output directory exists
output_dir = "./doisLR"
os.makedirs(output_dir, exist_ok=True)

paper_links = [
    "10.1016/j.jhtm.2022.04.008",
    "10.1108/IJCHM-01-2022-0142",
    "10.1111/emre.12561",
    "10.1108/ER-07-2021-0333",
    "10.1016/j.jbusres.2021.11.074"
]

paper_type = "doi"
proxies = {
    'http': 'socks5://127.0.0.1:7890'
}

session = requests.Session()
session.proxies = proxies

# Iterate over paper links with a progress bar
for paper_link in tqdm(paper_links, desc="Downloading papers"):
    # Extract the value after the last '/'
    paper_id = paper_link.split("/")[-1]
    out = f"{output_dir}/{paper_id}.pdf"

    headers = {'User-Agent': 'Mozilla/5.0'}
    response = session.get(f"https://sci-hub.se/{paper_link}", headers=headers)
    paper_url = response.url

    try:
        scihub_download(paper_url, paper_type=paper_type, out=out)
    except Exception as e:
        print(f"Could not download paper at {paper_link}: {e}")
```
